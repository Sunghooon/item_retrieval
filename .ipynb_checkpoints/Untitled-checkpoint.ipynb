```python
import cv2
import numpy as np
from tqdm import tqdm
from matplotlib import pyplot as plt
import pandas as pd
import torch
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#DATA_PATH = '../../../Documents/shopee-product-matching/'
DATA_PATH = 'shopee-product-matching/'

data = pd.read_csv(DATA_PATH + 'train.csv')
#train = pd.read_csv(DATA_PATH + 'train.csv')
train = data.head(250)
train['image'] = DATA_PATH + 'train_images/' + train['image']
test = data.loc[251:261]
test['image'] = DATA_PATH + 'train_images/' + test['image']
```

```python
# Train 데이터셋에서 특징 행렬 생성
def compute_feature_matrix(image_path_list):
    orb = cv2.ORB_create()
    feature_matrix = []
    
    images = [cv2.imread(path, cv2.IMREAD_GRAYSCALE) for path in image_path_list]
    keypoints_and_descriptors = [orb.detectAndCompute(image, None) for image in tqdm(images)]

    return keypoints_and_descriptors
```

```python
# 주어진 이미지에 대한 특징 행렬 생성
def compute_query_features(img_path):
    image = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
    orb = cv2.ORB_create()
    descriptors = orb.detectAndCompute(image, None)[1]
    return descriptors
```

```python
# 유사도 계산
def compute_similarity(query_features, feature_matrix):
    similarities = []
    for features in feature_matrix:
        similarity = cosine_similarity(query_features, features)
        #similarities.append(similarity)
        #print(similarity.shape)

    return similarities
```

```python
# 주어진 이미지가 어떤 이미지에 가장 유사한지 추론
def infer(query_image, train_feature_matrix):
    
    tmp_box = []
    #print(query_image)
    query_features = compute_query_features(query_image) 
    
    for keypoints, features_descriptors in train_feature_matrix:
        #print(features_descriptors)
        #print(query_features)
        #features_descriptors = features_descriptors.reshape(1, -1)
        similarity = cosine_similarity(query_features, features_descriptors)
        avg = np.mean(similarity.max(axis=1))
        tmp_box.append(avg)

    # 가장 유사한 이미지 인덱스 찾기
    #print(similarities)
    #most_similar_index = np.argmax(similarities)
    
    max_similarity = np.max(tmp_box)
    max_index = np.argmax(tmp_box)
    return max_similarity, max_index
```

```python
def train_img(train_images):
    train_feature_matrix = compute_feature_matrix(train_images)
    
    return train_feature_matrix
```

```python
train_img_list = train['image']
print('training img...')
train_feature_matrix = train_img(train_img_list)
```

```python
#print((train_feature_matrix[0][1]))
test_img_list = test['image']
#print(test_img_list)
most_similar_box = []
for i, test_img in enumerate(test_img_list):
    #print(query_img)
    max_val, max_idx = infer(test_img, train_feature_matrix)
    
    # Initiate ORB detector
    orb = cv2.ORB_create()
    query_img = cv2.imread(test_img)
    predicted_img = cv2.imread(train['image'][max_idx])
    keypoints1, descriptors1 = orb.detectAndCompute(query_img, None)
    keypoints2, descriptors2 = orb.detectAndCompute(predicted_img, None)
    # Create BFMatcher (Brute-Force Matcher) object
    bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True)
    # Match descriptors
    matches = bf.match(descriptors1, descriptors2)

    # Sort them in ascending order of distance
    matches = sorted(matches, key=lambda x: x.distance)

    # Draw first 10 matches (you can adjust the number)
    img_matches = cv2.drawMatches(query_img, keypoints1, predicted_img, keypoints2, matches[:10], None, flags=cv2.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS)
    print(test_img)
    print(train['image'][max_idx])
    cv2.imshow('Matches', img_matches)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    
    plt.subplot(1, 2, 1)
    plt.imshow(cv2.cvtColor(query_img, cv2.COLOR_BGR2RGB))
    plt.title("Image 1")

    plt.subplot(1, 2, 2)
    plt.imshow(cv2.cvtColor(predicted_img, cv2.COLOR_BGR2RGB))
    plt.title("Image 2")

    plt.show()
    
    most_similar_box.append([max_val, max_idx])
```
